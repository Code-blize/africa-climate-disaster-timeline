```python
columns_to_keep = ['Start Year', 'Disaster Type', 'Country', 'Total Affected', 'Total Deaths']

df_clean = df[columns_to_keep]
df_clean.shape
```

```python
df_clean.isnull().sum()
```

```python
df_clean['Disaster Type'].value_counts()
```

```python
climate_type = ['Flood', 'Storm', 'Drought', 'Wildfire', 'Extreme temperature']

df_climate = df_clean[df_clean['Disaster Type'].isin(climate_type)].copy()
df_climate.shape
```

```python
df_climate['Start Year'].value_counts().sort_index()
```

```python
df_climate['Start Year'].min(), df_climate['Start Year'].max()
```

Note: Raw data contains 1,883 climate disaster records after filtering. Downloaded from EM-DAT on April 16, 2026. Minor row differences from earlier analysis reflect live database updates.

```python
df_climate.isnull().sum()
```

```python
# Step 4: Aggregate by year
yearly = df_climate.groupby(['Start Year', 'Disaster Type']).agg(
    Events=('Disaster Type', 'count'),
    Total_Affected=('Total Affected', 'sum'),
    Total_Deaths=('Total Deaths', 'sum')
).reset_index()

yearly.head(10)
```

```python
yearly.shape
```

```python
yearly['Disaster Type'].value_counts()
```

```python
yearly['Total Affected'] = yearly['Total_Affected'].fillna(0).astype(int)
yearly['Total Deaths'] = yearly['Total_Deaths'].fillna(0).astype(int)
yearly.drop(columns=['Total_Affected', 'Total_Deaths'], inplace=True)
yearly.head()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
yearly.to_csv('africa_climate_timeline.csv', index=False)
```

```python
# Step 6: Pivot to wide format
yearly_wide = yearly.pivot(index='Start Year',
                           columns='Disaster Type',
                           values='Events').fillna(0).astype(int).reset_index()

yearly_wide.head()
```

```python
yearly_wide.to_csv('/content/drive/MyDrive/africa_climate_wide.csv', index=False)
```

```python
df_climate['Decade'] = (df_climate["Start Year"] // 10) * 10

df_climate[["Start Year", "Decade"]].head(10)
```

```python
df_climate.groupby(['Decade', 'Disaster Type']).agg(Event_count=('Disaster Type', 'count'))
```

```python
decade_summary = df_climate.groupby(['Decade', 'Disaster Type']).agg(
    Event_count=('Disaster Type', 'count'),
).reset_index()

decade_summary
```

```python
Decade_wide = decade_summary.pivot(index= 'Decade',
                                   columns = 'Disaster Type',
                                   values= 'Event_count').fillna(0).astype(int).reset_index()

Decade_wide
```

```python
Decade_wide.to_csv('/content/drive/MyDrive/africa_flood_dominance.csv', index=False)
```
